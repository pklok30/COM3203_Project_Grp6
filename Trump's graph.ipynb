# The process of making the graphs for Trump's tweets

Visualize the data we get from Mr. Donald Trump's tweets

```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Trump(processed).csv")
```

```python
#delete the id column
df.drop('id', axis=1, inplace=True)
```

```python
#delete the date column
df.drop('date', axis=1, inplace=True)
```

```python
#show all the data, set 60 because it is more than the total tweets number
df.head(60)
```

```python
#drop all the rows which contain NaN value
df.dropna()
```

```python
# count the numbers of type (column)
Type_count_series = df.groupby('type').size()
```

```python
# output a list which shows the frequencies of Fake news or Fact
for Type in Type_count_series.index:
    print(Type, Type_count_series[Type], sep=', total:')
```

```python
#output the mean of comments, retweets, likes for fake news and fact
df.groupby('type').mean()
```

# Bar chart and Pie chart for Retweets

```python
#group retweets by type
grouped_Retweets_Type = df["retweets"].groupby(df["type"])
```

```python
#draw a bar chart for the mean of retweets
grouped_Retweets_Type.mean().plot.bar(x='type', y='retweets',title='Mean of retweets (Trump tweets)')
```

```python
#draw a pie chart for the mean of retweets
grouped_Retweets_Type.mean().plot.pie(x="type", y="retweets",title='Mean of retweets (Trump tweets)')
```

# Bar chart and Pie chart for Comments

```python
#group comments by type
grouped_Comments_Type = df["comments"].groupby(df["type"])
```

```python
#draw a bar chart for the mean of comments
grouped_Comments_Type.mean().plot.bar(x='type', y='comments',title='Mean of comments (Trump tweets)')
```

```python
#draw a pie chart for the mean of comments
grouped_Comments_Type.mean().plot.pie(x="type",y="comments",title='Mean of comments (Trump tweets)')
```

# Bar chart and Pie chart for Likes

```python
#group likes by type
grouped_Likes_Type = df["likes"].groupby(df["type"])
```

```python
#draw a bar chart for the mean of likes
grouped_Likes_Type.mean().plot.bar(x='type', y='likes',title='Mean of likes (Trump tweets)')
```

```python
#draw a pie chart for the mean of likes
grouped_Likes_Type.mean().plot.pie(x="type", y="likes",title='Mean of likes (Trump tweets)')
```
